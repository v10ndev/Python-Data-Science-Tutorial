### Fancy Indexing

```python
arr2d = np.zeros((10,10))
```

```python
arr_length = arr2d.shape[1]
```

```python
for i in range(arr_length):
    arr2d[i] = i
    
arr2d
```

```python
arr2d[[2,4,6,8]]
```

```python
arr2d[[6,4,2,7]]
```
